```python
# PyMuPDF
import fitz
import os
```

```python
# PDF 폴더 설정
pdf_folder = "pdf"
output_folder = "pdf_images"
os.makedirs(output_folder, exist_ok=True)
```

```python
# 파일 정렬
pdf_files = sorted([f for f in os.listdir(pdf_folder) if f.endswith(".pdf")])
```

```python
# 문제 번호 시작
problem_number = 1
```

```python
for pdf_filename in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_filename)
    doc = fitz.open(pdf_path)

    # 문제 번호 범위 계산
    start_num = problem_number
    end_num = problem_number + 9

    # 1페이지 (개요) 저장
    cover_page = doc.load_page(0)
    cover_pix = cover_page.get_pixmap(dpi=200)
    cover_filename = f"문제{start_num}_{end_num}개요.png"
    cover_path = os.path.join(output_folder, cover_filename)
    cover_pix.save(cover_path)
    print(f"Saved: {cover_filename}")

    # 2~11페이지 (문제 10개) 저장
    for i in range(1, 11):
        page = doc.load_page(i)
        pix = page.get_pixmap(dpi=200)

        filename = f"문제{problem_number}번.png"
        filepath = os.path.join(output_folder, filename)
        pix.save(filepath)
        print(f"Saved: {filename}")

        problem_number += 1
```
