# 문제4. Mjob과 Fjob열에 capitalizer 함수 적용된 결과 출력

```python
df_sac[["Mjob", "Fjob"]].map(capitalizer)
```

# 문제5. age가 17을 초과하는지 여부를 True, False로 표현하는 legal_drinker 열 생성

```python
def majority(x):
    return x > 17
```

```python
df_sac["legal_drinker"] = df_sac["age"].map(majority)
```

```python
df_sac.head()
```

```python
df_sac["legal_drinker"] = df_sac["age"].map(lambda x: x > 17)
```

```python
df_sac.head()
```

# 문제6. int타입의 데이터에만 10을 곱하기

```python
def times10(x):
    if type(x) == int:
        return 10 * x
    else:
        return x
```

```python
df_sac.map(times10)
```

```python
df_sac.map(lambda x: 10 * x if type(x) == int else x)
```
