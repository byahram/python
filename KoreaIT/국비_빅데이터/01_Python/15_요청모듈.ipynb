# **requests 로 웹페이지 접속**

```python
url = "https://www.naver.com/"

rq.get(url)
```

```python
rq.post(url)
```

```python
url = "https://www.naver.com/a"

res = rq.get(url)

# 응답코드 가져오기
print(res.status_code)
```

```python
url = "https://www.naver.com/"

res = rq.get(url)

print(res.status_code)
```

```python
# 헤더 가져오기
print(res)
print(res.headers)
```

```python
# html 코드 보기
print(res.text)
```

```python
# 바이너리 형태로 값을 가져오기 때문에 res.text 에서 한글이 깨지는 현상을 방지할 수 있음
print(res.content)
```

```python
# 인코딩 확인
res.encoding
```

```python
# 쿼리스트링 데이터 생성
res = rq.get(url, params = {"key1" : "value1", "key2" : "value2"})

print(res.url)
```

# **urllib 으로 웹페이지 접속**

```python
from urllib.request import urlopen
```

```python
url = "https://www.naver.com/"

page = urlopen(url)

print(page)
```

```python
print(page.code)
print(page.headers)
print(page.url)
print(page.info().get_content_charset())
```

```python
# requests의 content() 함수와 동일한 기능
print(page.read())
```
